```python
# Attrition_Flag, Gender를 레이블 인코딩 해준다.
```

```python
le = LabelEncoder()

df['Attrition_Flag'] = le.fit_transform(df['Attrition_Flag'])
df['Gender'] = le.fit_transform(df['Gender'])
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
# 피처와 레이블 분리
y = df['Attrition_Flag']
X = df.drop('Attrition_Flag', axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2024)
```

---

```python
# RandomForest 사용
RFC = RandomForestClassifier(random_state=42)
RFC.fit(X_train, y_train)
y_pred = RFC.predict(X_test)

print(RFC.__class__.__name__," : ",accuracy_score(y_test, y_pred))
```

---

```python
# XGBoost 사용
XGB = XGBClassifier(random_state=42)
XGB.fit(X_train, y_train)
y_pred = XGB.predict(X_test)

print(XGB.__class__.__name__," : ",accuracy_score(y_test, y_pred))
```
