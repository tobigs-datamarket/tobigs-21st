<h1>결측치, 이상치 확인 및 처리</h1>

```python
all_columns = data.columns.tolist()
num_columns = data._get_numeric_data().columns.tolist()
cat_columns = list(set(all_columns) - set(num_columns))

df=data
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
# 이상치 확인 및 처리 (IQR method)

for column in num_columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    
df.reset_index(drop=True, inplace=True)
```

```python
df
```

<h1>데이터 시각화</h1>

```python
fig, axes = plt.subplots(nrows=5, ncols=6, figsize=(30, 25))
axes = axes.flatten()

for i, column in enumerate(num_columns):
    sns.histplot(df[column], kde=True, ax=axes[i])
    axes[i].set_title(f'[Histogram] {column}', fontsize=10)
    axes[i].set_xlabel(column, fontsize=8)
    axes[i].set_ylabel('value', fontsize=8)

for i in range(len(num_columns), 30):
    axes[i].set_visible(False)

plt.tight_layout()
plt.show() 
```

```python
fig, axes = plt.subplots(3, 3, figsize=(15, 15))
axes = axes.flatten()

for i, column in enumerate(cat_columns[:9]):
    sns.countplot(x=column, data=df, ax=axes[i])
    axes[i].set_title(f'[CountPlot] - {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('value')
    axes[i].tick_params(labelrotation=45)

for i in range(len(cat_columns), 9):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```

시각화를 통해 모든 값이 동일한 칼럼이 있음을 확인하였고 아래 코드를 통해 이들을 제거하였다.

```python
df = df.drop(['EmployeeCount', 'PerformanceRating', 'StandardHours', 'Over18'], axis=1)
cat_columns = [column for column in cat_columns if column != 'Over18']

df
```

<h1>상관관계 파악</h1>

```python
num_columns = df._get_numeric_data().columns.tolist()

corr_matrix = df[num_columns].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("numeric data corr")
plt.show()
```
